```python
import keras
import numpy as np
import os

from keras.models import Model,load_model
from PIL import Image

#global params
T = 10;
N = 5;
units = 1000;
splitFactor = 2/3;


#-- Layer specific params
# popToLayer = 6; # has outdim 13x13x256
# denseLayerInput = (T, 43268); # 13x13x256 + 4
popToLayer = 0; #outDim 13x13x125
denseLayerInput = (T, 21129); # 13x13x125 + 4

#---- vars to store---
imgData = [];

#--------- loading data -----------
def loadData(categories):
    trainImgList = [];
    testImgList = [];
    trainBBoxList = [];
    testBBoxList = [];
    for categoryPath in categories:
        imgDirPath = os.path.join(categoryPath, 'img');
        gtFilePath = os.path.join(categoryPath, 'groundtruth_rect.txt');
        dirImages = os.listdir(imgDirPath);
        dirBoxes = [];
        file = open(gtFilePath, "r");
        
        index = 0;
        # extract bound rect and normalize
        # extract bound rect and normalize
        for line in file:
            # append image dir
            dirImages[index] = os.path.join(imgDirPath, dirImages[index]);

            # get size of images & store image data for use in lossfn.
            currImage = Image.open(dirImages[index]);
            image_data = np.array(currImage, dtype='float32');
            image_data /= 255.;
            imgData.append(image_data);
            width, height = currImage.size;
            
            # normalize bounding box
            currBox = line.rstrip().split('\t');
            if(len(currBox)==1):
                currBox = line.rstrip().split(',');
            currBox[0] = int(currBox[0])/width;currBox[2] = int(currBox[2])/width;
            currBox[1] = int(currBox[1])/height;currBox[3] = int(currBox[3])/height;
            dirBoxes.append(currBox);
            index += 1;
            
        splitIndex = int(len(dirImages)*splitFactor);
        print('split',splitIndex);
        trainImgList += dirImages[0:splitIndex];
        testImgList += dirImages[splitIndex:];
        
        trainBBoxList += dirBoxes[0:splitIndex];
        testBBoxList += dirBoxes[splitIndex:];
    return (trainImgList, testImgList, trainBBoxList, testBBoxList);
        
#--------- YOLO Utils ------------
def loadYOLOModel(path):
    pretrainedModel = load_model('yolo-voc.h5');
    for i in range(0, popToLayer):
        pretrainedModel.layers.pop();

    x = pretrainedModel.layers[-1].output; #[13,13,256]
    return Model(input=pretrainedModel.input,output=x);

def _getYOLOFeat(model, image):
    model_image_size = model.layers[0].input_shape[1:3];
    resized_image = image.resize(
        tuple(reversed(model_image_size)), Image.BICUBIC)
    image_data = np.array(resized_image, dtype='float32');
    image_data /= 255.;
    image_data = np.expand_dims(image_data, 0);  # Add batch dimension.
    return model.predict(image_data);
    
def getYOLOFeats(model, imagePathList):
    featList = [];
    #print('paths',imagePathList);
    for inpImage in imagePathList:
        imageData = Image.open(inpImage);
        feat = _getYOLOFeat(model, imageData);
#       print('feat_shape:', feat.shape);
        featList.append(feat);
    return np.array(featList);

#---------- Extract data & model --------
yoloModel = loadYOLOModel('yolo-voc.h5');
(trainImgList, testImgList, trainBBoxList, testBBoxList) = loadData(['Human9']);
```

```python
# import keras
import numpy as np

from keras import backend as K
from keras.models import Sequential, Model
from keras.layers import Dense, LSTM, Lambda, BatchNormalization
from keras.layers.advanced_activations import LeakyReLU

seed = 10;
N = 5;
currGrpImgaData = [];
sigma = 0.01;

#-------- Metrics ---------------

def reshape_loss_inputs(groundList, predList, size):
    return (K.reshape(groundList, (size,4)), K.reshape(predList, (size,4)));
            
def avg_max_diff(groundList, predList):
    (groundList, predList)  = reshape_loss_inputs(groundList, predList, T);
    reward_loss = 0;
    for index in range(0,T):
        ground = groundList[index];
        pred = predList[index];
        (imwidth, imheight) = currGrpImgaData[index].size;
        imdata = currGrpImgaData[index];
            
        ground_x_min = ground[0]*imwidth;
        ground_y_min = ground[1]*imheight;
        ground_x_max = ground_x_min + K.maximum(0.0, ground[2])*imwidth;
     
        ground_y_max = ground_y_min + K.maximum(0.0, ground[3])*imheight;
        
        pred_x_min = pred[0]*imwidth;
        pred_y_min = pred[1]*imheight;
        pred_x_max = pred_x_min + K.maximum(0.0, pred[2])*imwidth;
        pred_y_max = pred_y_min + K.maximum(0.0, pred[3])*imheight;
        
def compute_squared_loss(groundList, predList):
    loss = 0;
    (groundList, predList)  = reshape_loss_inputs(groundList, predList, T);
    rewardSum = 0;
    for index in range(0,T):
        ground = groundList[index];
        pred = predList[index];
        # compute l(t)_1:N
        pred_x = K.random_normal([N], pred[0], sigma, seed=seed);
        pred_y = K.random_normal([N], pred[1], sigma, seed=seed);
        pred_w = K.random_normal([N], pred[2], sigma, seed=seed);
        pred_h = K.random_normal([N], pred[3], sigma, seed=seed);
        iterRewards = [];
        mean_reward = 0; #bs
        for iter in range(0,N):
            squaredErr = K.pow(- ground[0] + pred_x[iter], 2) + K.pow(- ground[1] + pred_y[iter], 2);
            squaredErr += K.pow(- ground[2] + pred_w[iter], 2) + K.pow(- ground[3] + pred_h[iter], 2);
            iterRewards.append(squaredErr);
            mean_reward += squaredErr;
        mean_reward /= N;
        mean_reward = np.ones(N)*mean_reward;
        iterRewards = iterRewards-mean_reward;
        rewardSum += iterRewards;
    return (rewardSum);

def compute_iou(groundList, predList):
    K.print_tensor(predList, "prediction is: ");
#     print('ground: ',groundList);
#     print('pred: ',predList);
    (groundList, predList)  = reshape_loss_inputs(groundList, predList, T);
    reward_loss = 0;
    for iter in range(0,N):  
        for index in range(0,T):
            ground = groundList[index];
            pred = predList[index];
            ground_x_min = ground[0];
            ground_y_min = ground[1];
            ground_x_max = ground_x_min + K.maximum(0.0, ground[2]);
            ground_y_max = ground_y_min + K.maximum(0.0, ground[3]);

            pred_x_min = pred[0];
            pred_y_min = pred[1];
            pred_x_max = pred_x_min + pred[2];
            pred_y_max = pred_y_min + pred[3];

            max_ymin = K.maximum(ground_y_min, pred_y_min);
            min_ymax = K.minimum(ground_y_max, pred_y_max);
            heights_intersect = K.maximum(0.0, min_ymax - max_ymin);

            max_xmin = K.maximum(ground_x_min, pred_x_min);
            min_xmax = K.minimum(ground_x_max, pred_x_max);
            width_intersect = K.maximum(0.0, min_xmax - max_xmin);

            intersection =  heights_intersect*width_intersect;
            intersection = K.maximum(0.0, intersection);
            K.print_tensor(intersection, "intersection is: ");

            area_pred = (pred_y_max - pred_y_min)*(pred_x_max - pred_x_min);
            area_ground = (ground_y_max - ground_y_min)*(ground_x_max - ground_x_min);
            union = K.maximum(0.0, area_pred + area_ground - intersection);
            reward = intersection/union;

            reward_loss += (1-reward);
            K.print_tensor(area_pred, "pred area: ");
            #print('intersection',intersection);
    K.print_tensor(reward_loss/N, "reward loss: ");

    return reward_loss/N;
    
    
#--------- Utils ------------
def getMuFromH(x):
    return x[:,:,units-4:units];

#-------- DeepRL Model ----------
def getDRLModel():
    model = Sequential();
    model.add(Dense(units, activation= 'linear', name='rnnFeat', input_shape = denseLayerInput)); 
    #model.add(BatchNormalization());
#     model.add(LeakyReLU(alpha=0.1)) ;
    model.add(LSTM(units, return_sequences = True)); # input [1 T units]
    model.add(Lambda(getMuFromH)); # output - (1, timesteps, 4)
    return model;

#--------- Main func --------------
drlModel = getDRLModel();
adam = keras.optimizers.Adam(lr=0.00001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
drlModel.compile(loss=compute_squared_loss,
              optimizer=adam);
print(drlModel.summary()); 


currEpoch = 1;
numIters = 3;

for itr in range(0, numIters): # 10 iters
#     if(itr == numIters/2):
#         drlModel.compile(loss=compute_iou,
#               optimizer=adam);

    for imgListIter in range(0, int(len(trainImgList)/T)):
        stIndex = imgListIter*T;
        imgGrp = trainImgList[stIndex:stIndex+T];
        boxGrp = np.array(trainBBoxList[stIndex:stIndex+T], dtype='float32' );
        imgGrpYOLOFeatList = getYOLOFeats(yoloModel, imgGrp);
        imgGrpFeatList = [];
        for index,imgFeat in enumerate(imgGrpYOLOFeatList):
            imgFeat = np.reshape(imgFeat, [-1, ]); #flatten out image feat

            if(index==0):
                imgGrpFeatList.append(np.append(imgFeat, boxGrp[index])); #concat location to first frame
            else:
                imgGrpFeatList.append(np.append(imgFeat, [0,0,0,0]));
        imgGrpFeatList = np.array(imgGrpFeatList);
        imgGrpFeatList = np.reshape(imgGrpFeatList, (1, T, -1));
        print('imgrp shape:', imgGrpFeatList.shape);
        boxGrp = np.reshape(boxGrp, (1, T, 4));
        #imgDatacurrGrpImgaData = np.array(imgData[(imgListIter-1)*T : imgListIter*T]);
        #--- Fit model ---------
        print('epoch is: ', currEpoch);
        history = drlModel.fit(imgGrpFeatList, boxGrp, batch_size = 1, epochs=1);
        #print('history',history);
        currEpoch += 1;

# y = drlModel.predict(imgGrpFeatList);
#weights = drlModel.get_weights();

# print("last pred y is: ",y);
# print("last gt val: ",boxGrp);
```
